````python
# Hint
# plt.legend(), plt.title(), plt.scatter()  

def show2D(pca_score, Y, target_names):
    plt.figure()
    # TODO -->
    size = len(target_names)
    plt.title('Result')
    for i in range(size):
        plt.scatter(pca_score[Y == i,0], pca_score[Y == i,1], marker='o', s=30)
    plt.legend(target_names, loc='lower right')
    # <--
````

````python
show2D(PC_score, Y, target_names)
````

# TODO 2

PCA  .

## 
1. normalize  
2.  numpy    
    - ```numpy.mean()```
    - ```numpy.linalg.eig()```
    - ```numpy.dot()```

````python
def student_pca(X, n_components=2):
    pca_results = None
    # TODO -->
    # preprocessing
    X = X.T
    for i in range(len(X)):
        X[i] = X[i] - np.mean(X[i])
    X = X.T
    cov_X = np.cov(X.T)
    
    # eigen val, vec
    eigen_value, eigen_vector = np.linalg.eig(cov_X)
    
    pca_results = np.dot(eigen_vector.T, X.T)
    pca_results = pca_results.T[:, :2]
    # <--
    return pca_results

pca_student_score = student_pca(X, n_components=2)
````

````python
show2D(pca_student_score, Y, target_names)
````
